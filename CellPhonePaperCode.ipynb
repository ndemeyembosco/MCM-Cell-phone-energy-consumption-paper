```python
# find voltage and milliamp of battery
import re

voltage = []
milliamp = []

for b in table1.ix[:,2]:
    
    name = re.findall(r'[A-Z][a-z]+', b)
    mp = re.findall(r'[0-9]+', b)
    
    if "Ni" in name:
        voltage.append(1.2)
    elif "Li" in name:
        if "Ion" in name:
            voltage.append(3.6)
        elif "Po" in name:
            voltage.append(3)
    if len(mp) != 0:
        milliamp.append(int(mp[0]))
    else:
        milliamp.append(0)
            
print("Voltage:", voltage)
print("Milliamp:", milliamp)
```

```python
# calculate Wh
Wh = []
for i in range(len(voltage)):
    watt = (milliamp[i] * voltage[i]) / 1000
    Wh.append(watt)
print(Wh)
```

```python
class Household():
    
    states = {"LAND": 0, "TRANSIT": 1, "CELL": 2}
    
    def __init__(self):
        self.member = 3 # Average people in a US household is 3 
        self.landline = 1 # We are assuming that every household has only one landline
        self.cell = 0
        self.state = "LAND"
    
    def getLandline(self):
        return self.landline
    
    def getCell(self):
        return self.cell
#        self.index = index
    
#    def update(self):
#        for m in range(self.member):
#            prob = np.random.random()
#            if prob < self.index:
#                self.cell += 1
#                self.state = "TRANSIT"
#                if self.cell == self.member:
#                    self.state == "CELL"
#                    self.landline = 0
#        return self.landline, self.cell, self.state
    
    def update(self):
        for e in range(3):
            self.cell += 1
            if self.state == "LAND":
                self.state = "TRANSIT"
            if self.cell == self.member:
                self.landline = 0
                self.state = "CELL"
            if self.cell > 3:
                self.cell = 3
        return self.landline, self.cell, self.state
        
```

```python
## MAIN CODE

popularity_index = -1.0
counter = 0

# list to plot
x = []
land = []
ce = []
y = []


household_list = [Household() for i in range(round(household[0]).astype(int))]
prev_household = 0
l = []
c = []

# for every household
for i in range(len(household)):
    # set temp variables
    total_l = 0
    total_c = 0
    total_s = 0
    # add more household
    difference = round(household[i]) - len(household_list)
    for n in range(difference.astype(int)):
        household_list.append(Household())
    
    for h in household_list:
        rand = np.random.random()
        if rand < popularity_index:
            l, c, s = h.update()
            total_l += l
            total_c += c
        else:
            total_l += h.getLandline()
            total_c += h.getCell()
################## UNTIL HERE IT WORKS! ####################
    current_year = year[i]
    if table1.ix[:,0][counter] == current_year:
        counter += 1 # use counter - 1
        
    landlinewatts = total_l * 2.5 * 24
    cellwatts = total_c * Wh[counter - 1] * 2
    
    x.append(current_year)
    y.append(landlinewatts + cellwatts)
    ce.append(cellwatts)
    land.append(landlinewatts)
    
fig = plt.gcf()
ax = fig.add_subplot(111)
plt.plot(x,y, label = "Total Energy Consumption")
plt.plot(x,ce, ':d', label = "Cellphone Energy Consumption")
plt.plot(x,land, '--',label = "Landline Energy Consumption")
plt.xlabel("Years")
plt.ylabel("MW per day")
plt.title("Energy Consumption of the US")
handles, labels = ax.get_legend_handles_labels()
leg = ax.legend(handles, labels, loc = 'upper center', bbox_to_anchor = (0.5,1))
fig.set_size_inches(12, 9)
fig.savefig('ElectricConsump05.png', dpi=100)
plt.legend()

    
    
    
    
    


    
    
```
