### Differential Equations

```python
## dL/dt = - a * H
## dT/dt = a * H - b* T
## dC/dt = b * T
#  dH/dt = 1.3  # slope of the line of best fit of growth of population

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from scipy.integrate import odeint


a = 0.3
b = 0.1
c = 0.01
def cellRev(u, t):
    L, T, C, H = u
    dL =  - a *L
    dT = (a*L - b*T)
    dC = ( b*T + c*H)
    dH = 1.3
    return [dL, dT, dC, dH]

xs = range(1983, 2015)
ys = odeint(cellRev, [83.93, 0, 0, 83.93], xs)
# plt.plot(xs, ys[:,3], label="Households")
fig = plt.gcf()
ax= fig.add_subplot(111)
plt.title("Differential equations model of The cell phone revolution")
plt.ylabel("Number of households (millions)")
plt.xlabel("Time (years)")
plt.plot(xs, ys[:,2], label="Cells only")
plt.plot(xs, ys[:,1], label="Transitioning")
plt.plot(xs, ys[:,0], label="Landline only")
handles,labels = ax.get_legend_handles_labels()
leg = ax.legend(handles,labels,loc = 'upper center', bbox_to_anchor = (0.5,1))
fig.set_size_inches(12,9)
fig.savefig('Differential1.png')
plt.legend()
```

```python
voltages = [0.36, 0.456, 0.6, 1.8, 1.5, 3.6, 3.06, 4.5, 3.492, 4.392, 4.824, 5.112, 5.94, 7.56, 10.08, 9.36, 12.0]
years = [1983, 1992, 1996, 1998, 1999, 2000, 2003, 2005, 2006, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
new_voltages = []
k = list(zip(voltages, years))
for i in range(len(k)-1):
    n = k[i + 1][1] - k[i][1]
    if n > 1:
        for j in range(n):
            new_voltages.append(k[i][0])
    else:
        new_voltages.append(k[i][0])

energys_cells = [i*j for i,j in zip(new_voltages, ys[:, 2])]
energys_landlines = [2.5*j for j in ys[:, 0]]

```

```python
fig = plt.gcf()
ax= fig.add_subplot(111)
plt.title("Differential equations model of The cell phone revolution")
plt.ylabel("Number of households (millions)")
plt.xlabel("Time (years)")
plt.plot(xs, energys_cells, label="Cells only")
plt.plot(xs, energys_landlines, label="Landline only")
handles,labels = ax.get_legend_handles_labels()
leg = ax.legend(handles,labels,loc = 'upper center', bbox_to_anchor = (0.5,1))
fig.set_size_inches(12,9)
fig.savefig('Differential3.png')
plt.legend()
```
